# Converting categorical values into numerical values

```python
#First find all the unique values for each object column
def count_unique_values(data):
    objects = data.select_dtypes(include=['object']).columns
    for col in objects:
        unique = data[col].value_counts()
        print(unique, '\n')
        print('*'*50)
count_unique_values(df)
```

```python
df.columns
```

```python
# Category Mappings

object_columns = [
    'loan_limit', 'approv_in_adv', 'loan_type', 'loan_purpose',
    'Credit_Worthiness', 'open_credit', 'business_or_commercial',
    'Neg_ammortization', 'interest_only', 'lump_sum_payment',
    'construction_type', 'occupancy_type', 'Secured_by', 'total_units',
    'credit_type', 'co-applicant_credit_type', 'age',
    'submission_of_application', 'Security_Type'
]

# Perform one-hot encoding for each object column
for col in object_columns:
    # Get one-hot encoding for the current column
    one_hot_encoded = pd.get_dummies(df[col], prefix=col)
    # Concatenate the one-hot encoded DataFrame with the original DataFrame
    df = pd.concat([df, one_hot_encoded], axis=1)
    # Drop the original column after one-hot encoding
    df.drop(columns=[col], inplace=True)





# Display the updated DataFrame with one-hot encoded values
print(df)
```

```python
def count_features(data):
    #Getting column with objects dtype
    objects= data.select_dtypes(include=['object']).columns
    numerical=data.select_dtypes(include=['int64']).columns
    floating=data.select_dtypes(include=['float']).columns
    
    print('Numerical features are :',len(numerical))
    print('Objects features are  :',len(objects))
    print('Floating features are :',len(floating))
    
count_features(df)
```

```python
# Training model
X=df.drop('Status',axis='columns')
Y=df.Status
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.30, random_state = True)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
```

```python
from sklearn.linear_model import LogisticRegression
logistic_regression = LogisticRegression(penalty='l1', solver='liblinear')
logistic_regression.fit(X_train, Y_train)
Y_predicted_logistic = logistic_regression.predict(X_test)
print(accuracy_score(Y_test,Y_predicted_logistic))
print(precision_score(Y_test,Y_predicted_logistic))
```

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier(max_depth = 5) 
decision_tree.fit(X_train, Y_train)
Y_predicted_decisionTree = decision_tree.predict(X_test)
accuracy_dt = accuracy_score(Y_test,Y_predicted_decisionTree)
precision_dt = precision_score(Y_test,Y_predicted_decisionTree)
print("Acuuracy : ", accuracy_dt)
print("Precision : ", precision_dt)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, Y_train)
Y_pred_rf = rf.predict(X_test)
accuracy_rf = accuracy_score(Y_test, Y_pred_rf)
precision_rf = precision_score(Y_test,Y_predicted_decisionTree)
print("Accuracy:", accuracy_rf)
print("Precision : ", precision_rf)
```
